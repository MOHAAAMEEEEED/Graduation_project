## Preprocessing

```python
import nltk
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.corpus import wordnet

from nltk.stem import WordNetLemmatizer
import re

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')

stop_words = set(stopwords.words('english'))
translated_table = str.maketrans('', '', string.punctuation)
```

```python
def get_wordnet_pos(tag):
    """Map POS tag to WordNet POS tag."""
    tag_map = {
        'J': wordnet.ADJ,
        'V': wordnet.VERB,
        'N': wordnet.NOUN,
        'R': wordnet.ADV
    }
    return tag_map.get(tag[0], wordnet.NOUN)
```

```python
def preprocess_text(text):
    """
    Preprocess text by performing lowercasing, removing numbers,
    punctuation, stopwords, and lemmatization.
    """
    if not isinstance(text, str):
        return ""
    
    # Initialize resources
    stop_words = set(stopwords.words('english'))
    lemmatizer = WordNetLemmatizer()
    
    # Preprocessing steps
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    
    # Tokenization and filtering
    tokens = word_tokenize(text)
    filtered_words = [word for word in tokens if word not in stop_words]
    
    # Lemmatization with POS tagging
    lemmatized_words = []
    for word in filtered_words:
        pos_tag = nltk.pos_tag([word])[0][1]
        wordnet_pos = get_wordnet_pos(pos_tag)
        lemma = lemmatizer.lemmatize(word, pos=wordnet_pos)
        lemmatized_words.append(lemma)
    
    return ' '.join(lemmatized_words)
```

```python
# Apply preprocessing to all text columns
for column in df_sub.columns:
    df_sub[column] = df_sub[column].astype(str).apply(preprocess_text)

df_sub.head()
```

## Pipeline for computing similarity

#### TF-IDF

```python
import torch
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sentence_transformers import SentenceTransformer
```

```python
def compute_tfidf_similarity(df):
    """Compute TF-IDF similarity between job description and responses."""
    similarities = []
    for _, row in df.iterrows():
        texts = [
            row["Job Description"],
            row["Resp1 (20)"],
            row["Resp2 (40)"],
            row["Resp3 (60)"],
            row["Resp4 (80)"],
            row["Resp5 (100)"]
        ]
        vectorizer = TfidfVectorizer()
        tfidf_matrix = vectorizer.fit_transform(texts)
        cosine_sim = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:])
        similarities.append(cosine_sim.flatten().tolist())
    return similarities
```

#### Jaccard

```python
import textdistance as td
def compute_jaccard_similarity(df):
    """Compute Jaccard similarity between job description and responses."""
    similarities = []
    for _, row in df.iterrows():
        job_desc = row["Job Description"]
        response_similarities = []
        
        for resp_col in ["Resp1 (20)", "Resp2 (40)", "Resp3 (60)", 
                        "Resp4 (80)", "Resp5 (100)"]:
            response = row[resp_col]
            jaccard_score = td.jaccard.similarity(
                set(job_desc.split()),
                set(response.split())
            )
            response_similarities.append(jaccard_score)
        
        similarities.append(response_similarities)
    return similarities
```

#### Compute Transformer-Based Similarity (BERT & RoBERTa & DistilBERT )

```python
def compute_transformer_similarity(df, model_name):
    """Compute transformer-based similarity between job description and responses."""
    model = SentenceTransformer(model_name)
    similarities = []
    
    for _, row in df.iterrows():
        texts = [
            row["Job Description"],
            row["Resp1 (20)"],
            row["Resp2 (40)"],
            row["Resp3 (60)"],
            row["Resp4 (80)"],
            row["Resp5 (100)"]
        ]
        embeddings = model.encode(texts, convert_to_tensor=True)
        cosine_sim = torch.nn.functional.cosine_similarity(
            embeddings[0].unsqueeze(0),
            embeddings[1:],
            dim=1
        )
        similarities.append(cosine_sim.tolist())
    
    return similarities
```

```python
df_sub["TF-IDF Similarity"] = compute_tfidf_similarity(df_sub)
df_sub["Jaccard Similarity"] = compute_jaccard_similarity(df_sub)

# Compute transformer-based similarities
df_sub["DistilBERT Similarity"] = compute_transformer_similarity(
    df_sub, "sentence-transformers/distilbert-base-nli-mean-tokens"
)
df_sub["BERT Similarity"] = compute_transformer_similarity(
    df_sub, "sentence-transformers/all-MiniLM-L6-v2"
)
df_sub["RoBERTa Similarity"] = compute_transformer_similarity(
    df_sub, "sentence-transformers/all-distilroberta-v1"
)
```

```python
import pandas as pd
import numpy as np

def display_full_similarities(df, similarity_columns):
    """
    Display complete similarity scores for each row.
    
    Args:
        df: DataFrame containing similarity scores
        similarity_columns: List of column names containing similarity scores
    """
    pd.set_option('display.max_rows', None)
    pd.set_option('display.max_columns', None)
    pd.set_option('display.width', None)
    pd.set_option('display.max_colwidth', None)
    
    for idx, row in df.iterrows():
        print(f"\nRow {idx}:")
        for col in similarity_columns:
            scores = row[col]
            print(f"\n{col}:")
            for i, score in enumerate(scores, 1):
                print(f"Response {i}: {score:.6f}")
        print("\n" + "="*50)

# List of similarity columns
similarity_columns = [
    'TF-IDF Similarity',
    'Jaccard Similarity',
    'DistilBERT Similarity',
    'BERT Similarity',
    'RoBERTa Similarity'
]

# Display all similarity scores
display_full_similarities(df_sub, similarity_columns)
```
